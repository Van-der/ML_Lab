```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Seed for reproducibility
np.random.seed(42)

# Generate temperatures between 10 and 40 degrees Celsius (100 samples)
temperature = np.random.uniform(10, 40, 100)

# Generate number of ice creams sold using a polynomial relationship plus noise
# Formula: sales = a * temp^2 + b * temp + c + noise
a = 2
b = 10
c = 50
noise = np.random.normal(0, 20, 100)

sales = a * temperature**2 + b * temperature + c + noise

# Create a DataFrame
df = pd.DataFrame({
    'Temperature': temperature,
    'IceCreams_Sold': sales
})

# Save to CSV
df.to_csv('icecream_sales_polyreg.csv', index=False)

# Optional: Plot to visualize data
plt.scatter(df['Temperature'], df['IceCreams_Sold'], color='blue', alpha=0.6)
plt.xlabel('Temperature (°C)')
plt.ylabel('Ice Creams Sold')
plt.title('Temperature vs Ice Creams Sold (Polynomial Relationship)')
plt.show()

print("Dataset 'icecream_sales_polyreg.csv' created.")
print(df.head())
```
