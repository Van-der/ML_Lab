```python
from sklearn.datasets import load_iris
from scipy.stats import mode
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score
import numpy as np  
```

```python
iris=load_iris()
X=iris.data
y=iris.target #y do we not reshape y?
kmeans=KMeans(n_clusters=3)
kmeans.fit(X)
y_kmeans=kmeans.predict(X)
```

```python
labels=np.zeros_like(y_kmeans)
for i in range(3):
    mask=(y_kmeans==i)
    labels[mask]=mode(y[mask])[0]

accuracy=accuracy_score(y,labels)
print(f'Accuracy: {accuracy*100:.2f}%')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=labels, palette='viridis')
sns.scatterplot(x=kmeans.cluster_centers_[:, 0], y=kmeans.cluster_centers_[:, 1], s=200, marker='X', color='red', label='Centroids')
plt.title("KMeans Clustering of Iris data")
plt.show()
```
